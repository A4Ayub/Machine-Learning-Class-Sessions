Observations and Decisions
1. The columns thrust, timings and size have over 80% records missing so we choose to drop the columns
2. The distance column have m characters, we need to remove these characters and fill the null values with the mode
3. Eye power has null values which we need to replace with the mode

```python
working_dataframe_df["thrust"].value_counts()
```

```python
working_dataframe_df["timings"].value_counts()
```

```python
working_dataframe_df["size"].value_counts()
```

```python
# drop the columns
working_dataframe_df.drop(["thrust","timings","size"],axis=1,inplace=True)
```

```python
working_dataframe_df
```

```python
working_dataframe_df.info()
```

```python
# convert the distance column to string just to be sure
convertion_dict_03 = {"distance":str}
working_dataframe_df = working_dataframe_df.astype(convertion_dict_03)
```

```python
# The distance column has some values that are null, others have characters in prefix 
# so the objective is to remove the characters, temporarily replace the NULL string with 0, convert the datatype to int
# Replace the zero with the mode
working_dataframe_df["distance"] = working_dataframe_df["distance"].map(lambda x: x.lstrip("").rstrip("m"))
working_dataframe_df
```

```python
# Replace the null strings with zero
working_dataframe_df["distance"] = working_dataframe_df["distance"].replace(to_replace=["NULL"],value="0")
```

```python
working_dataframe_df
```

```python
# convert the distance column back to int
convertion_dict_04 = {"distance":int}
working_dataframe_df = working_dataframe_df.astype(convertion_dict_04)
```

```python
working_dataframe_df.info()
```

```python
# replace the zero values with the mode
distance_mean = working_dataframe_df["distance"].mean()
distance_mean
```

```python
working_dataframe_df["distance"] = working_dataframe_df["distance"].replace(to_replace=0,value=distance_mean)
working_dataframe_df
```

```python
working_dataframe_df["distance"] = working_dataframe_df["distance"].round(2)
```

```python
working_dataframe_df
```

```python
# eye_power has some null string values so we need to replace those with the mode
working_dataframe_df["eye_power"].value_counts()
```

```python
working_dataframe_df["eye_power"] = working_dataframe_df["eye_power"].replace(to_replace=["NULL"],value="s")
```
